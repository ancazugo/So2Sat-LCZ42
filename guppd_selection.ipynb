```python
import os
from dotenv import load_dotenv
import pandas as pd
import geopandas as gpd

load_dotenv()
```

```python
guppd_path = os.getenv("GUPPD_PATH")
so2sat_path = os.getenv("SO2SAT_DIR")
guppd_gdf = gpd.read_file(guppd_path, layer="urbanspatial_guppd_v1_polygons")
guppd_gdf.sort_values(by="GHSPOP_R23_2025", inplace=True)
guppd_gdf = guppd_gdf[guppd_gdf["GHSPOP_R23_2025"] > 1e5]
```

```python
cities_00_bounds_lst = ["Bogota", "Buenos Aires", "Cairo", "Caracas", 
                        "Chicago", "Dhaka", "Guangzhou", "Hong Kong",
                        "Jakarta", "Karachi", "Lima", "New York",
                        "Philadelphia", "Quezon City [Manila]", 
                        "Salvador", "Tokyo"]
cities_01_bounds_lst = ["Lisbon", "Los Angeles", "Nairobi", "Rio De Janeiro",
                        "San Jose", "Santiago", "Shanghai", "São Paulo",
                        "Tehran", "Vancouver", "Washington D.C."]
cities_02_bounds_lst = ["Changsha", "Nanjing", "Osaka [Kyoto]", "Rome",
                        "Sydney"]
cities_03_bounds_lst = ["Berlin", "Cape Town", "Cologne", "Milan", "Moscow",
                        "Mumbai", "Paris", "Rawalpindi [Islamabad]"]
cities_04_bounds_lst = ["Amsterdam", "Beijing", "Munich", "Qingdao", "Wuhan"]
cities_05_bounds_lst = ["Istanbul", "London", "Madrid", "Melbourne", "Zurich",
                        "东营区"]
```

```python
from shapely.geometry import box

# Create a copy of the GeoDataFrame
so2sat_guppd_bounds_gdf = so2sat_guppd_gdf.copy()

# Get the bounds of each geometry: this returns a DataFrame with columns 'minx', 'miny', 'maxx', 'maxy'
bounds_df = so2sat_guppd_bounds_gdf['geometry'].bounds

# Replace the geometry column with bounding box polygons
so2sat_guppd_bounds_gdf['geometry'] = bounds_df.apply(lambda row: box(row['minx'], row['miny'], row['maxx'], row['maxy']), axis=1)

# Assign the buffer size for each list
city_buffer_sizes = {}

for city in cities_00_bounds_lst:
    city_buffer_sizes[city] = 0.0
for city in cities_01_bounds_lst:
    city_buffer_sizes[city] = 0.1
for city in cities_02_bounds_lst:
    city_buffer_sizes[city] = 0.2
for city in cities_03_bounds_lst:
    city_buffer_sizes[city] = 0.3
for city in cities_04_bounds_lst:
    city_buffer_sizes[city] = 0.4
for city in cities_05_bounds_lst:
    city_buffer_sizes[city] = 0.5

# Helper to get city buffer by name
def get_city_buffer(name):
    return city_buffer_sizes.get(name, 0.0)

# You may need to normalize city names if different, example:
# so2sat_guppd_bounds_gdf["JRC_NAME_MAIN"].unique()

# Apply buffer to geometry based on city list
def buffer_row(row):
    city_name = row["JRC_NAME_MAIN"]
    buffer_dist = get_city_buffer(city_name)
    return row["geometry"].buffer(buffer_dist)

so2sat_guppd_bounds_gdf["geometry"] = so2sat_guppd_bounds_gdf.apply(buffer_row, axis=1)

# Add each column to the GeoDataFrame
bounds_df = so2sat_guppd_bounds_gdf['geometry'].bounds
for col in bounds_df.columns:
    so2sat_guppd_bounds_gdf[col] = bounds_df[col].values

so2sat_guppd_bounds_gdf.to_file(so2sat_path + "so2sat_guppd_bounds_gdf.gpkg", index=False)
so2sat_guppd_bounds_gdf[['SMOD_ID', 'JRC_NAME_MAIN', 'minx', 'miny', 'maxx', 'maxy']].to_csv(so2sat_path + "so2sat_guppd_bounds_gdf.csv", index=False)
```
